# Degree vs no degree for business owners
Use CBO dataset

```python
# selecting relevant rows
degrees_of_owners = cleaned_cbo[
    (cleaned_cbo['OWNER_SEX'] == '001') & 
    (cleaned_cbo['NAICS2017'] == '00') &
    (cleaned_cbo['OWNER_RACE'] == '00') &
    (cleaned_cbo['OWNER_ETH'] == '001') & 
    (cleaned_cbo['OWNER_VET'] == '001') &
    (cleaned_cbo['QDESC'] == 'O08')
    ]
degrees_of_owners
```

```python
# checking datatypes
degrees_of_owners.info()
```

```python
# changing datatypes from object to numeric
degrees_of_owners = degrees_of_owners.astype({'OWNPDEMP':'int', 'OWNPDEMP_PCT':'float'})
degrees_of_owners.info()
```

```python
# selecting columns and sorting
degrees_of_owners = degrees_of_owners[['OWNCHAR', 'OWNCHAR_LABEL', 'OWNPDEMP', 'OWNPDEMP_PCT']]
degrees_of_owners
```

```python
# selecting row and plotting

# selecting row
degrees_plot_table = degrees_of_owners.loc[[127098, 127100]]

# plotting
degrees_plot = degrees_plot_table.plot(
    title = "Degrees of Business Owners",
    kind = "bar",
    legend = False
)

degrees_plot.set_xticklabels(['Some degree', 'No degree'])
degrees_plot.set_yticks([0e6, 0.5e6, 1e6, 1.5e6, 2e6, 2.5e6])
degrees_plot.set_yticklabels([0, 0.5, 1, 1.5, 2, 2.5])

degrees_plot.set_xlabel('\nSource: Annual Business Survey 2018: Characteristics of Business Owners')
degrees_plot.set_ylabel('Number of respondents in millions')
plt.show()
```

# Company Revenue by Owner Race
Using CB dataset

```python
# selecting a slice to see owners race without duplication, dropping unnecessary columns
revenue_race = cleaned_cb[
    (cleaned_cb['BUSCHAR'] == 'A1') & 
    (cleaned_cb['QDESC'] == 'B01') &
    (cleaned_cb['SEX'] == '001') &
    (cleaned_cb['ETH_GROUP'] == '001') &
    (cleaned_cb['VET_GROUP'] == '001') &
    (cleaned_cb['NAICS2017'] == '00')
    ]

revenue_race = revenue_race.drop([
    'SEX', 'SEX_LABEL', 
    'ETH_GROUP', 'ETH_GROUP_LABEL', 
    'VET_GROUP', 'VET_GROUP_LABEL',
    'NAICS2017', 'NAICS2017_LABEL',
    'EMP', 'EMP_F',	
    'EMP_PCT', 'EMP_PCT_F',	
    'PAYANN', 'PAYANN_F',	
    'PAYANN_PCT', 'PAYANN_PCT_F',
    'RCPPDEMP_F', 'RCPPDEMP_PCT', 'RCPPDEMP_PCT_F',
    'FIRMPDEMP_F', 'FIRMPDEMP_PCT', 'FIRMPDEMP_PCT_F',
    'QDESC', 'QDESC_LABEL',
    'BUSCHAR', 'BUSCHAR_LABEL'
    ], axis = 1)

revenue_race = revenue_race.astype({'FIRMPDEMP':'int64', 'RCPPDEMP':'int64'})

revenue_race
```

```python
# calculating average revenue, dropping uncessary rows
revenue_race['AVGREV'] = (revenue_race['RCPPDEMP']/revenue_race['FIRMPDEMP'])

labels_to_include = ['White', 'Black or African American', 'American Indian and Alaska Native', 'Asian', 'Native Hawaiian and Other Pacific Islander']

revenue_race = revenue_race[revenue_race['RACE_GROUP_LABEL'].isin(labels_to_include)].drop(['RACE_GROUP', 'FIRMPDEMP', 'RCPPDEMP'], axis = 1).sort_values(by = 'AVGREV', ascending = False)

revenue_race
```

```python
# plotting graph
revenue_race_plot = revenue_race.plot(
    title = 'Average Revenue per US Firm by Owner Race',
    legend = False,
    kind = 'bar',
    grid = True
)

# cleaning xtick labels
revenue_race_plot.set_xticklabels([
    'White',
    'American Indian/Alaska Native',
    'Asian',
    'Native Hawaiian/Other Pacific Islander',
    'Black/African American'
])

# axes labels
revenue_race_plot.set_xlabel('Source: Annual Business Survey 2018: Characteristics of Businesses')
revenue_race_plot.set_ylabel('Average Revenue Per Firm (in $1000)')

plt.show()
```
